# **Importing Libraries and Dataset**

```python
import pandas as pd
import numpy as np
from datetime import date


data=pd.read_excel('Assignment_Timecard (1).xlsx')
data.head(10)
```

# **Converting the Time column to Datatime Format And Extracting the day from that**

```python
data['Time'] = pd.to_datetime(data['Time'], errors='coerce')
data.info()
data['Time'].dt.day
data['date']=data['Time'].dt.day
data.iloc[1:20]
```

# **Selecting only few columns on basis of which we are going find the result**

```python
data=data[['Position ID','Employee Name','date']]
data.head()
```

# **Creating a new dataframe called new_df and Adding a new column in it called 'difference' to store the difference between two consecutive values in 'date' column**

```python
new_df=pd.DataFrame(data)
new_df['difference']=new_df['date'].diff()
new_df['difference']=new_df['difference'].fillna(value=0)
new_df['difference']=new_df['difference'].astype(int)
new_df = new_df[new_df['difference'] != 0]
```

# **Defining a Function to iterate over the difference column and whenever it encounter 6 consecutive 1's(p.s  it means that particular person has worked for 7 consecutive days) it will replace the first encountered 1 with a random number(e.g. 100). Whenever this function is called it will take a column(i.e. list) as argument and it will return an updated column(i.e. list).**

```python
def replace_first_1_with_100(column):
    count = 0
    for i in range(len(column)):
        if column.iloc[i] == 1:
            count += 1
            if count == 6:
                column.iloc[i-5] = 100
                break
        else:
            count = 0
    return column
```

# **1. Grouping the new_df on the basis of 'Employee Name' and applying the above function to the 'difference' columns of each groups individually and storing the result in a variable called grouped_ones**
#**2. Now adding the grouped_one list to the dataframe new_df as a column called 'ones'.**
#**3. Filtering all the values in 'ones' which are 100 and their corresponding 'Employee Name' and 'Position ID'.**
#**4.Printing all the 'Employee Name' and 'Position ID' who have worked for 7 consecutive days**

```python
grouped_ones=new_df.groupby('Employee Name')['difference'].apply(replace_first_1_with_100)
new_df['ones']= grouped_ones
result_df = new_df[new_df['ones']== 100]
result_df=result_df[['Position ID','Employee Name']]
print("Employees who has worked for 7 consecutive days:")
print(" ")
print(result_df)
```
